## Explain the distance metric you utilized to calculate the similarity/dissimilarity between small molecules.

I initially used a jaccard index to calculate distance between all molecules. However, in my KMeans implementation I used a euclidean distance on the distance-space embedding to cluster. 

## Use a dimensionality reduction algorithm (PCA, t-SNE, UMAP, etc) to generate a 2D visualization of the small molecule dataset. Each point should represent asingle molecule. (Note: you may have to plot a subset of your data, depending on which dimensionality reduction algorithm you choose.)

```python
from umap import UMAP

ligands = Ligand("../data/subset.csv")
distmat = ligands.pdist(metric='jaccard')

um = UMAP()
scatter = pd.DataFrame(
    um.fit_transform(distmat), 
    columns = ['x', 'y']
)
sns.scatterplot(data=scatter, x='x', y='y')
```

## Cluster the small molecules using your implementation of a partitioning clustering algorithm. Visualize this clustering by coloring clusters on the 2D visualization generated in question 2.

```python
km = PartitionClustering(metric='euclidean',seed=3)
labels = km.cluster(distmat, k=7)

scatter['km_labels'] = labels
sns.scatterplot(data=scatter, x='x', y='y', hue='km_labels', palette="Set1")
```

## Explain your choice of partitioning clustering algorithm. Is it sensitive to initialization conditions? How do you select the number of clusters?

I implemented a K-Means Clustering algorithm. It is sensitive to initialization conditions - but after implementing K++ Initialization it is less sensitive to sampling issues. The number of clusters can be selected either by selecting a minima of global distance as calculated by the algorithm or by calculating silhouette coefficients and selecting the condition which maximises the average silhouette coefficient. 

## Cluster the small molecules using your implementation of a hierarchical clustering algorithm. Visualize this clustering in the same way as question 3.

```python
hc = HierarchicalClustering()
zmat = hc.cluster(distmat)
lineage = hc.get_lineage()
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram

plt.figure(figsize=(10,10))
dendrogram(zmat)
plt.show()
```

```python
idx_break = np.flatnonzero(zmat[:,2] > 0.02).min()
scatter['hc_labels'] = lineage[idx_break]
sns.scatterplot(data=scatter, x='x', y='y', hue='hc_labels')
```

## Explain your choice of hierarchical clustering algorithm. Is it sensitive to initialization conditions? How do you select the number of clusters?

I implemented an agglomerative clustering algorithm. Observations are iteratively joined to their nearest neighbors until all observations belong to a single cluster. It is not sensitive to initialization conditions. The number of clusters are selected by thresholding the number of generations away from the root of the tree to allow for clusters. This can be measured with silhouette scores.  

## Evaluate the quality of both clusterings using your implementation of a clustering quality metric. Explain your choice of quality metric. Which clustering performed ‘best’ according to your metric?

```python
print("Silhouette Coefficient K-Means : {}".format(km.score().mean()))
print("Silhouette Coefficient HClust : {}".format(hc.score(idx_break).mean()))
```

K-Means performed considerably better, I believe this is due to the fact that the lineage output of hclust resulted in a very step-like lineage so there was no meaningful clustering beyond the identical samples.

## Compare the two clusterings using your implementation of clustering similarity. How similar are the two clusterings using this function?

```python
similarities = ClusterSimilarity(labels, lineage[idx_break])
print("Similarity Index : {}".format(similarities))
```

The two clusters are ~20% similar for each observation, this is likely due to the number of identical fingerprints that are clustered identically between the two methods.

## For the “best” clustering, as determined by your quality metric, visualize the distribution of Autodock Vina scores in each cluster. Do members of the same cluster have similar docking scores? Why or why not?

```python
scatter['autodock_score'] = ligands.frame.Score.values

sns.boxplot(
    data=scatter, x='km_labels', y='autodock_score'
)
```

Yes the members cluster similarly, but there are outliers. A reason this may happen is from collisions in the fingerprint calculation (i.e. multiple features hashing to the same bit). Another reason this could happen is poor clustering. What is interesting is that the cluster with identical activations has the highest amount of variance which leads me to think that introducing features beyond the scaffold could result in more meaningful clusters.

## Select the top scoring molecule from each cluster. This is your list of cluster heads. Visualize the top 5 by score in PyMOL and pick your favorite. Are they structurally diverse?

```python
import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem

from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole

sorted_SMILES = ligands.frame.sort_values("Score").SMILES.values

mols = [Chem.MolFromSmiles(s) for s in sorted_SMILES[:5]]
Draw.MolsToGridImage(mols, molsPerRow=5, subImgSize=(200, 200))
```

They are structurally diverse, but whats interesting about them are the frequency of hydroxyl groups on them. This is a great example of how fingerprints can fail to capture molecular features because they don't generally capture the number of times a feature is seen - just whether or not it was found. Depending on how the fingerprint was created though - parameterizing the radius, as well as changing from bitvector to count vector - it's possible to capture this.
